```python
import networkx as nx
%matplotlib inline

import matplotlib.pyplot as plt
```

```python
#create a an empty graph with no nodes and no edges

G = nx.Graph()
```

```python
#add a single node

G.add_node(1)
```

```python
#add a list of nodes

G.add_nodes_from([2, 3])
```

```python
nx.draw_networkx(G, node_color='green', node_size=700)
```

# Add Edges

```python
#adding edges to a graph
G.add_edge(1,2)
```

```python
nx.draw_networkx(G)
```

```python
#add edges using unpack edge tuples

e = (2,3)
G.add_edge(*e)
```

```python
nx.draw_networkx(G)
```

```python
#adding list of edges
G.add_edges_from([(1,2),(1,3)])
```

```python
nx.draw_networkx(G)
```

```python
#find number of nodes in a graph
print (G.number_of_nodes())
```

```python
#find number of egdes in a graph
print (G.number_of_edges())
```

# Getting Graph Properties

## get some properties of a graph

```python
#getting the number of nodes is a graph
print(G.nodes())
```

```python
#getting the number of edges in a graph
print(G.edges())
```

```python
#getting the neighbor of a connected graph
print(G.neighbors(1))
```

```python
# get the degrees from a graph
print(G.degree())
```

```python
# find if a graph has a an edge
print(G.has_edge(2,3))
```

```python
# find if a graph has a an edge
print(G.has_edge(2,4))
```

```python
# find if there exist a node in a graph
print(G.has_node(4))
```

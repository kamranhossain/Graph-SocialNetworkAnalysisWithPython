```python
import networkx as nx
%matplotlib inline

import matplotlib.pyplot as plt
```

```python
#create a an empty graph with no nodes and no edges

G = nx.Graph()
```

```python
#add a single node

G.add_node(1)
```

```python
#add a list of nodes

G.add_nodes_from([2, 3])
```

```python
nx.draw_networkx(G, node_color='green', node_size=700)
```

# Add Edges

```python
#adding edges to a graph
G.add_edge(1,2)
```

```python
nx.draw_networkx(G)
```

```python
#add edges using unpack edge tuples

e = (2,3)
G.add_edge(*e)
```

```python
nx.draw_networkx(G)
```

```python
#adding list of edges
G.add_edges_from([(1,2),(1,3)])
```

```python
nx.draw_networkx(G)
```

```python
#find number of nodes in a graph
print (G.number_of_nodes())
```

```python
#find number of egdes in a graph
print (G.number_of_edges())
```

# Node Manipulation

```python
H = nx.Graph()
```

```python
edgelist = [(0,1),(1,2),(2,3),(0,3)]
```

```python
H.add_edges_from(edgelist)
```

```python
nx.draw(H,with_labels=True, node_color='cyan',node_size=700)
```

```python
# remove a node
H.remove_node(3)
nx.draw(H,with_labels=True, node_color='cyan',node_size=700)
```

```python
# remove a list of nodes
H.remove_nodes_from([0,1])
nx.draw(H,with_labels=True, node_color='cyan',node_size=700)
```

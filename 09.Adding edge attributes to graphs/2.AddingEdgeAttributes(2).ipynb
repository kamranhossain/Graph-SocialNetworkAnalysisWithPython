# Adding edge attributes to our graph

```python
import networkx as nx
%matplotlib inline

import matplotlib.pyplot as plt
```

```python
H = nx.Graph()
```

```python
# adding edges to the graph
H.add_edge(1,2, weight=4.7)
```

```python
H.add_edges_from([(3,4),(4,5)],color='Cyan')
```

```python
H.add_edges_from([(1,2,{'color':'blue'}),(2,3,{'weight':8})])
```

```python
nx.draw_networkx(H,with_labels=True, node_color='green', node_size=900)
```

```python
H[2][3]['weight'] = 4.7
```

```python
#view edge properties
H.edges(data = True)
```

# Adding edge attributes to our graph 2

```python
cities = nx.Graph()
```

```python
cities.add_edge('San Diago', 'L.A',{'distance':0.4})
```

```python
cities.add_edge('New York', 'Nashville',{'distance':5.6})
```

```python
cities.add_edge('Boston','DC',{'distance':0.8})
```

```python
cities.edges(data=True)
```

```python
nx.draw(cities,node_color='#9999FF', node_size=800, with_labels=True)
```

# Creating Directed Graph

```python
import networkx as nx
%matplotlib inline

import matplotlib.pyplot as plt
```

```python
b = nx.DiGraph([{'a','b'},{'b','c'},{'c','a'}])
```

```python
#adding egdes to di graph
b.add_edge('a','b')

b.add_edges_from([{'d','c'},{'d','b'}])
```

```python
#plot our graph
nx.draw_networkx(b,with_labels=True, node_size=800,node_color='cyan')
```

```python
# print the out-degrees of the di-graph
print(b.out_degree())
```

```python
#print the in-degrees of the di-graph
print(b.in_degree())
```

```python
#print some other properties of the graph
print(b.in_edges())
```

# More on DiGraphs

```python
dg = nx.DiGraph()
```

```python
dg.add_edges_from([(1,2),(1,3),(2,4),(2,5),(2,6),(3,7),(7,8)])
```

```python
nx.draw(dg,node_color='#9999FF',with_labels=True, node_size=750)
```

```python
print(nx.degree_centrality(dg))
```

```python
print(nx.betweenness_centrality(dg))
```

```python
print(nx.closeness_centrality(dg))
```

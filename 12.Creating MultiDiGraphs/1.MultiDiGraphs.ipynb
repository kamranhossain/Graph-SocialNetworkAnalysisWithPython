# Multi DiGraphs

```python
import networkx as nx
%matplotlib inline

import matplotlib.pyplot as plt
```

```python
G = nx.MultiDiGraph()
```

```python
G.add_node(1)
G.add_nodes_from([2,3])
G.add_nodes_from(range(100,110))
nx.draw(G,node_color='#A0CBE2',node_size=800,with_labels=True)
```

```python
GM = nx.MultiDiGraph(day='Sunday', month='January',Year='2015', DOB='02/14/1985')
```

```python
GM.graph
```

```python
GM.add_node(1, time='11:03pm ')
```

```python
GM.node[1]
```

```python
GM.add_edge(1,2, weight=100)
```

```python
GM.add_edges_from([(3,4),(4,5)], color='green')
```

```python
nx.draw(GM, with_labels=True, node_color='c',node_size=2000)
```

```python
print(GM.nodes(data=True))
```

```python
print(GM.edges(data=True))
```

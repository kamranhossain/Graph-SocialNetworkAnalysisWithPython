# Graph Generator 01

```python
import networkx as nx
%matplotlib inline

import matplotlib.pyplot as plt
```

```python
# draw a complete_graph
plt.figure(figsize=(10,7))
cg = nx.complete_graph(5, create_using=nx.DiGraph())
nx.draw_networkx(cg)
```

```python
print(cg.in_degree())
```

```python
print(cg.out_degree())
```

```python
#draw a graph called path-graph
plt.figure(figsize=(10,7))
pg=nx.path_graph(5,create_using=nx.DiGraph())
nx.draw_networkx(pg)
```

```python
# draw a gnr-graph, which is like a facebook or twitter
plt.figure(figsize=(10,7))
gnr = nx.gnr_graph(10,1,create_using=nx.DiGraph())
nx.draw_networkx(gnr)
```

```python
print(gnr.in_degree())
```

```python
# draw a gnr-graph, which is like a facebook or twitter
plt.figure(figsize=(10,7))
gnr = nx.gnr_graph(100,1,create_using=nx.DiGraph())
nx.draw_networkx(gnr)
```

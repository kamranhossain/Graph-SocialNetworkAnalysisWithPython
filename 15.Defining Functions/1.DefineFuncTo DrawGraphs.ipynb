# Basic drawing using python functions

```python
import networkx as nx
%matplotlib inline

import matplotlib.pyplot as plt
```

```python
# define a function in Python
def draw_graph():
    G = nx.Graph()
    G.add_edges_from([(1,2),(2,3),(1,3),(1,4)])
    nx.draw(G,node_size=800,node_color='cyan')
```

```python
# call the draw_graph function
draw_graph()
```

```python
# define a function in Python
def draw_graph():
    G = nx.Graph()
    G.add_edges_from([(1,2),(2,3),(1,3),(1,4)])
    nx.draw(G,node_size=800,node_color='cyan',with_labels=True)
```

```python
draw_graph()
```

```python
#define a bew function call draw_barabasi_albert
def draw_barabasi_albert():
    G = nx.barabasi_albert_graph(100,10)
    hst = nx.degree_histogram(G)
    plt.hist(hst, bins=20,color='red')
```

```python
draw_barabasi_albert()
```

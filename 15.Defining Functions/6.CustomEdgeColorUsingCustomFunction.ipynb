### Customize node size and node colors in the Graph

```python
import networkx as nx
%matplotlib inline

import matplotlib.pyplot as plt
```

```python
def custom_node_size_and_node_color():
    #create our empty graph
    G = nx.Graph()
    
    #add edges 
    G.add_edge('A','B')
    G.add_edge('B','C')
    G.add_edge('C','A')
    
    ## draw the graph
    nx.draw(G,with_labels=True)
    plt.show()
    
    
    #define the nodelist
    nodelst = ['A','B','C']
    
    #define the colorlist
    colorlst = ['c','m','y']
    
    #define the nodesizelist
    nodesize = [1000,2000,3000]
    
    nx.draw(G,with_labels=True, node_color=colorlst, node_size=nodesize,nodelist=nodelst)
    
    plt.show()
```

```python
custom_node_size_and_node_color()
```

### Customize edge colors in the Graph

```python
def custom_edge_colors():
    G = nx.Graph()
    
    #add edges 
    G.add_edge('A','B')
    G.add_edge('B','C')
    G.add_edge('C','A')
    
    ## draw the graph
    nx.draw(G,with_labels=True)
    plt.show()
    
    
    #create list of edge color
    edgecol = ['c','y','m']
    
    #define the nodesizelist
    nodesize = [1000,2000,3000]
    
    #create the edge lsit
    edgelst = [('A','B'),('B','C'),('C','A')]

    nx.draw(G,edgelist=edgelst, edge_color=edgecol,with_labels=True)
    plt.show()
```

```python
# call the function
custom_edge_colors()
```
